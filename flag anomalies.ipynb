```python
import pandas as pd

fields = ['Id', 'Family Id (Family)', 'Snapshot Number', 'Longitude', 'Latitude', 'Age', 'Time Total Survey (In minutes)']
filename = 'database - extra data'
df = pd.read_excel(f'{filename}.xlsx', usecols=fields)
```

```python
df.info()
```

```python
df.sort_values(by=['Family Id (Family)', 'Snapshot Number', 'Age'], inplace=True, ignore_index=True)
df
```

```python
# add the flag column to keep track of flagged cases
# False means the survey is not flagged, and True means the survey is flagged for one of the reasons discussed in the Summary Report
df['flagged for invalid entries'] = False
df['reason(s) for invalid entries'] = ''
df
```

```python
# flag surveys that are geographically misplaced

df['Latitude'] = df['Latitude'].str.replace(',','.')
df['Longitude'] = df['Longitude'].str.replace(',','.')
df['invalid latitude value'] = False
df['invalid longitude value'] = False

index = 0
while index < len(df):
    try:
        df.at[index, 'Latitude'] = float(df.at[index, 'Latitude'])
    except:
        latitude = df.at[index, 'Latitude']
        df.at[index, 'flagged for invalid entries'] = True
        if (df.at[index, 'reason(s) for invalid entries'] == ''):
            df.at[index, 'reason(s) for invalid entries'] += (f'could not convert latitude value of {latitude} to float')
        else:
            df.at[index, 'reason(s) for invalid entries'] += (f', could not convert latitude value of {latitude} to float')
    
    try:
        df.at[index, 'Longitude'] = float(df.at[index, 'Longitude'])
    except:
        longitude = df.at[index, 'Longitude']
        df.at[index, 'flagged for invalid entries'] = True
        if (df.at[index, 'reason(s) for invalid entries'] == ''):
            df.at[index, 'reason(s) for invalid entries'] = (f'could not convert longitude value of {longitude} to float')
        else:
            df.at[index, 'reason(s) for invalid entries'] += (f', could not convert longitude value of {longitude} to float')

    index = index + 1
```

```python
from global_land_mask import globe
# global_land_mask library credit: Karin, Todd. Global Land Mask. October 5, 2020. https://doi.org/10.5281/zenodo.4066722

df['is land'] = True
index = 0
while index < len(df):
    if (not df.at[index, 'flagged for invalid entries']):
        latitude = df.at[index, 'Latitude']
        longitude = df.at[index, 'Longitude']
        if (longitude <= 180) and (longitude >= -180):
            if (latitude <= 90) and (latitude >= -90):
                try:
                    is_ocean = globe.is_ocean(latitude, longitude)
                    if (is_ocean):
                        df.at[index, 'flagged for invalid entries'] = True
                        df.at[index, 'is land'] = False
                        if (df.at[index, 'reason(s) for invalid entries'] == ''):
                            df.at[index, 'reason(s) for invalid entries'] += (f'({latitude}, {longitude}) value is in ocean')
                        else:
                            df.at[index, 'reason(s) for invalid entries'] += (f', ({latitude}, {longitude}) value is in ocean')
                except:
                    df.at[index, 'flagged for invalid entries'] = True
                    if (df.at[index, 'reason(s) for invalid entries'] == ''):
                        df.at[index, 'reason(s) for invalid entries'] += (f'error identifying whether the ({latitude}, {longitude}) values pair is on water or land')
                    else:
                        df.at[index, 'reason(s) for invalid entries'] += (f', error identifying whether the ({latitude}, {longitude}) values pair is on water or land')
            else:
                df.at[index, 'flagged for invalid entries'] = True
                if (df.at[index, 'reason(s) for invalid entries'] == ''):
                    df.at[index, 'reason(s) for invalid entries'] += (f'latitude value of {latitude} out of range [-90, 90]')
                else:
                    df.at[index, 'reason(s) for invalid entries'] += (f', latitude value of {latitude} out of range [-90, 90]')
        else:
            df.at[index, 'flagged for invalid entries'] = True
            if (df.at[index, 'reason(s) for invalid entries'] == ''):
                df.at[index, 'reason(s) for invalid entries'] += (f'longitude value of {longitude} out of range [-180, 180]')
            else:
                df.at[index, 'reason(s) for invalid entries'] += (f', longitude value of {longitude} out of range [-180, 180]')
    
    index = index + 1
```

```python
# read second excel file
import pandas as pd

fields = ['Id', 'Family Id', 'Survey Number', 'Date']
filename = 'database'
df = pd.read_excel(f'{filename}.xlsx', usecols=fields)
```

```python
# flag surveys that don't have valid snapshot values
import math

df['invalid snapshot value'] = False
df['invalid family id'] = False

index = 0
if (df.at[index, 'Family Id'] == ''):
    df.at[index, 'flagged for invalid entries'] = True
    df.at[index, 'invalid snapshot value'] = True
    df.at[index, 'invalid family id'] = True
    if (df.at[index, 'reason(s) for invalid entries'] == ''):
        df.at[index, 'reason(s) for invalid entries'] += ('family id value is missing')
    else:
        df.at[index, 'reason(s) for invalid entries'] += (', family id value is missing')

elif (df.at[index, 'Survey Number'] == 1):
    df.at[index, 'flagged for invalid entries'] = False
    df.at[index, 'invalid snapshot value'] = False

else:
    df.at[index, 'flagged for invalid entries'] = True
    df.at[index, 'invalid snapshot value'] = True
    if (df.at[index, 'reason(s) for invalid entries'] == ''):
        df.at[index, 'reason(s) for invalid entries'] += ('missing snapshot value of 1')
    else:
        df.at[index, 'reason(s) for invalid entries'] += (', missing snapshot value of 1')

index = index + 1
while index < len(df):
    if (math.isnan(df.at[index, 'Family Id'])):
        df.at[index, 'flagged for invalid entries'] = True
        df.at[index, 'invalid snapshot value'] = True
        df.at[index, 'invalid family id'] = True
        if (df.at[index, 'reason(s) for invalid entries'] == ''):
            df.at[index, 'reason(s) for invalid entries'] += ('family id value is missing')
        else:
            df.at[index, 'reason(s) for invalid entries'] += (', family id value is missing')

    else:
        if (df.at[index, 'Family Id'] == df.at[index - 1, 'Family Id']):
            
            if (df.at[index, 'Survey Number'] == (df.at[index - 1, 'Survey Number'] + 1)):
                df.at[index, 'flagged for invalid entries'] = False
                df.at[index, 'invalid snapshot value'] = False
            
            else:
                df.at[index, 'flagged for invalid entries'] = True
                df.at[index, 'invalid snapshot value'] = True
                previous_snapshot_id = df.at[index - 1, 'Id']
                if (df.at[index, 'Survey Number'] == (df.at[index - 1, 'Survey Number'])):
                    duplicated_snapshot_value = df.at[index, 'Survey Number']
                    if (df.at[index, 'reason(s) for invalid entries'] == ''):
                        df.at[index, 'reason(s) for invalid entries'] += (f'duplicated snapshot value of {duplicated_snapshot_value} with snapshot Id {previous_snapshot_id}')
                    else:
                        df.at[index, 'reason(s) for invalid entries'] += (f', duplicated snapshot value of {duplicated_snapshot_value} with snapshot Id {previous_snapshot_id}')
                
                if (df.at[index, 'Survey Number'] > (df.at[index - 1, 'Survey Number'] + 1)):
                    missing_snapshot_value = df.at[index, 'Survey Number'] + 1
                    if (df.at[index, 'reason(s) for invalid entries'] == ''):
                        df.at[index, 'reason(s) for invalid entries'] += (f'missing snapshot value of {missing_snapshot_value} after snapshot Id {previous_snapshot_id}')
                    else:
                        df.at[index, 'reason(s) for invalid entries'] += (f', missing snapshot value of {missing_snapshot_value} after snapshot Id {previous_snapshot_id}')
        
        else:
            
            if (df.at[index, 'Survey Number'] == 1):
                df.at[index, 'flagged for invalid entries'] = False
                df.at[index, 'invalid snapshot value'] = False
            
            else:
                df.at[index, 'flagged for invalid entries'] = True
                df.at[index, 'invalid snapshot value'] = True
                if (df.at[index, 'reason(s) for invalid entries'] == ''):
                    df.at[index, 'reason(s) for invalid entries'] += ('missing snapshot value of 1')
                else:
                    df.at[index, 'reason(s) for invalid entries'] += (', missing snapshot value of 1')
    index = index + 1
```

```python
# calculate time difference between baseline survey and follow-up one and flag the ones with less than 3 months time in difference for the same person
import numpy as np
pd.options.mode.chained_assignment = None

df['invalid date value'] = False
df['Date'] = pd.to_datetime(df['Date'])
last['Date'] = pd.to_datetime(last['Date'])
penultimate['Date'] = pd.to_datetime(penultimate['Date'])

last_family_ids = last['Family Id'].values.tolist()
penultimate_family_ids = penultimate['Family Id'].values.tolist()
```

```python
index = 0
total_time_difference_in_weeks = 0

while ((index < len(last)) and (index < len(penultimate))):
    if (last.at[index, 'Family Id'] == penultimate.at[index, 'Family Id']):
        last_date = last.at[index, 'Date']
        penultimate_date = penultimate.at[index, 'Date']
        time_difference = last_date - penultimate_date
        time_difference = time_difference / np.timedelta64(1, 'W')
        total_time_difference_in_weeks += time_difference
        if (time_difference < 12):
            original_index = last.at[index, 'index']
            df.at[original_index, 'invalid date value'] = True
            df.at[original_index, 'flagged for invalid entries'] = True
            if (df.at[index, 'reason(s) for invalid entries'] == ''):
                df.at[original_index, 'reason(s) for invalid entries'] += ('follow-up survey taken in less than 3 months')
            else:
                df.at[original_index, 'reason(s) for invalid entries'] += (', follow-up survey taken in less than 3 months')
    index = index + 1
```

```python
# calculate how many families have had follow-up surveys in less than 3 months
family_count_with_follow_ups = len(last['Family Id'].values.tolist())
invalid_date_value_count = df['invalid date value'].value_counts()[True]
print(f'the number of families that had taken a follow-up survey within less than three months is {invalid_date_value_count} out of {family_count_with_follow_ups} total families with follow-up surveys')
invalid_date_value_percentage = round((invalid_date_value_count / family_count_with_follow_ups) * 100, 2)
print(f'the percentage of families that had taken a follow-up survey within less than three months is {invalid_date_value_percentage}% out of {family_count_with_follow_ups} total families with follow-up surveys')
```

```python
# what is the average time between the baseline survey and the follow-up survey
family_count_with_follow_ups = len(last['Family Id'].values.tolist())
average = round(total_time_difference_in_weeks/4/family_count_with_follow_ups, 2)
print(f'the average time between the baseline survey and the follow-up survey is {average} months')
```

### Database - Indicators Excel file

```python
# additional analysis questions
# read indicators excel file
import pandas as pd

fields = ['Snapshot Id', 'Family Id', 'Survey Number', 'Survey Title', 'Stoplight Codename', 'Stoplight Value']
filename = 'database - indicators'
df = pd.read_excel(f'{filename}.xlsx', usecols=fields)
```

```python
df.info()
```

```python
df.sort_values(by=['Family Id', 'Stoplight Codename', 'Survey Number'], ascending = [True, True, False], inplace=True, ignore_index=True)
df
```

```python
# how many families have transitioned from yellow to green or from red to green per indicator
indicators = df['Stoplight Codename'].unique().tolist()
print(', '.join(indicators))
```

```python
df.info()
```

```python
# find last and penultimate survey taken for those families that took at least one follow-up survey
last_survey_number = pd.Series()
penultimate_survey_number = pd.Series()

index = 0
curr_family_id = df.at[index, 'Family Id']
curr_survey_number = df.at[index, 'Survey Number']
if (curr_survey_number >= 2):
    last_survey_number[curr_family_id] = curr_survey_number

prev_family_id = curr_family_id
prev_survey_number = curr_survey_number
index = index + 1

while (index < len(df)):
    curr_family_id = df.at[index, 'Family Id']
    curr_survey_number = df.at[index, 'Survey Number']

    if (curr_family_id == prev_family_id):
        if ((curr_survey_number < prev_survey_number) and (curr_family_id not in penultimate_survey_number)):
            penultimate_survey_number[curr_family_id] = curr_survey_number
    
    elif (curr_survey_number >= 2):
        last_survey_number[curr_family_id] = curr_survey_number

    prev_family_id = curr_family_id
    prev_survey_number = curr_survey_number
    index = index + 1
```

```python
# find last survey indices for each indicator taken for each family
last_survey_index = pd.Series()
penultimate_survey_index = pd.Series()

index = 0
curr_family_id = int(df.at[index, 'Family Id'])
curr_survey_number = int(df.at[index, 'Survey Number'])
if (curr_survey_number == last_survey_number[curr_family_id]):
    last_survey_index[curr_family_id] = pd.Series()
    last_survey_index[curr_family_id][df.at[index, 'Stoplight Codename']] = index

prev_family_id = curr_family_id
prev_survey_number = curr_survey_number
index = index + 1

while (index < len(df)):
    curr_family_id = int(df.at[index, 'Family Id'])
    curr_survey_number = int(df.at[index, 'Survey Number'])

    if (curr_family_id in last_survey_number):
        if (curr_survey_number == last_survey_number[curr_family_id]):
            stoplight_codename = df.at[index, 'Stoplight Codename']
            if (curr_family_id in last_survey_index):
                last_survey_index[curr_family_id][stoplight_codename] = index
            else:
                last_survey_index[curr_family_id] = pd.Series()
                last_survey_index[curr_family_id][stoplight_codename] = index

        elif (curr_survey_number == penultimate_survey_number[curr_family_id]):
            stoplight_codename = df.at[index, 'Stoplight Codename']
            if (curr_family_id in penultimate_survey_index):
                penultimate_survey_index[curr_family_id][stoplight_codename] = index
            else:
                penultimate_survey_index[curr_family_id] = pd.Series()
                penultimate_survey_index[curr_family_id][stoplight_codename] = index

        prev_family_id = curr_family_id
        prev_survey_number = curr_survey_number
    index = index + 1
```

```python
# calculate the number of transitions per indicator
indicator_transition_yellow_to_green = pd.Series()
indicator_transition_red_to_green = pd.Series()

family_ids = df['Family Id'].unique().tolist()
for family_id in family_ids:
    if ((family_id is not None) and (family_id in last_survey_number)):
        for indicator in indicators:
            if ((indicator in last_survey_index[family_id]) and (indicator in penultimate_survey_index[family_id])):
                last_indicator_value = df.at[last_survey_index[family_id][indicator], 'Stoplight Value']
                penultimate_indicator_value = df.at[penultimate_survey_index[family_id][indicator], 'Stoplight Value']

                if (penultimate_indicator_value is not None) and (last_indicator_value is not None):
                    transition = last_indicator_value - penultimate_indicator_value
                    if (transition == 2):
                        if (indicator in indicator_transition_red_to_green):
                            indicator_transition_red_to_green[indicator] += 1
                        else:
                            indicator_transition_red_to_green[indicator] = 1
                    
                    elif (transition == 1 and last_indicator_value == 3):
                        if (indicator in indicator_transition_yellow_to_green):
                            indicator_transition_yellow_to_green[indicator] += 1
                        else:
                            indicator_transition_yellow_to_green[indicator] = 1
```

```python
# print number of families that transitioned from yellow to green per indicator
total_num_of_families = len(family_ids)
print(f'number of familie that took at least one follow-up survey out of {total_num_of_families} and transitioned from yellow to green per indicator are:')
for indicator, count in indicator_transition_yellow_to_green.items():
    print(f'    indicator {indicator} had {count} transitions')
```
